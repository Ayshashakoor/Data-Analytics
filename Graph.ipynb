```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_set=pd.read_csv('Covid Data.csv')
print(data_set)
```

```python
import seaborn as sns
sns.countplot(x='breathing_issue', data=data_set)
plt.title('Breathing Issues')
plt.show()
```

```python
sns.histplot(data=data_set, x='breathing_issue', weights='Blood O2 Level in Percentage', bins=2, color='green')
plt.title('Breathing Issue Weighted by Blood O2 Level')
plt.show()
```

```python
data_set['Age Group'] = pd.cut(data_set['age'], bins=[0, 18, 40, 60, 80, 100], labels=['0-18', '19-40', '41-60', '61-80', '81+'])
sns.countplot(x='Age Group', hue='Needed Hospitalization', data=data_set, palette='muted')
plt.title('Age Group vs. Hospitalization')
plt.show()
```

Count the Occurence for Hospital

```python
hospital_counts = data_set['Needed Hospital'].value_counts()

# Plot the pie chart
plt.pie(hospital_counts, labels=['No Hospital', 'Need Hospital'], autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'red'])

# Add title
plt.title('Proportion of Hospitalization')
plt.show()
```
